```python
import pandas as pd
from textblob import TextBlob

df = pd.read_excel(r'C:\Users\mahat\Downloads\twitter_training.xlsx')

def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    return sentiment_score

df['SentimentScore'] = df['Text'].apply(analyze_sentiment)

df['SentimentScore'] = df['SentimentScore'].apply(lambda x: (x + 1) / 2)

df.to_excel(r'C:\Users\mahat\Downloads\twitter_training_analyzed.xlsx', index=False)
```
